```python
refe=open("C:/Users/RitKat/Desktop/future/ML/Baby'sName Data Scrapping/babynames/baby1990.html","r")
gg=refe.read()
```

```python
text=re.findall("<td>(\d+)</td><td>(\w+)</td><td>(\w+)</td>",gg)
```

```python
print(text)
```

```python
year=re.search("Popularity in (\d\d\d\d)",gg)
tot=[year.group(1)]
print(tot)
for x in text:
    m_name=x[1]+str(x[0])
    f_name=x[2]+str(x[0])
    tot.append(m_name)
    tot.append(f_name)
```

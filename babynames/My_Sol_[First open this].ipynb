```python
import re
```

```python
def format_it(address):
  refe=open(address,"r")
  gg=refe.read()
  text=re.findall("<td>(\d+)</td><td>(\w+)</td><td>(\w+)</td>",gg)
  year=re.search("Popularity in (\d\d\d\d)",gg)
  tot=[year.group(1)]
  for x in text:
      m_name=x[1]+str(x[0])
      f_name=x[2]+str(x[0])
      tot.append(m_name)
      tot.append(f_name)
  tot.sort()
  return tot
```

```python
def main():
  x=1990
  while x<2009:
    addr="C:/Users/RitKat/Desktop/Projects_Flash/Baby'sName Data Scrapping/babynames/baby"+str(x)+".html"
    listy=format_it(addr)
    print (listy)
    print (" ")
    x=x+2
    
```
